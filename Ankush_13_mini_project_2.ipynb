```python
import pandas as pd

file_path = r'D:\DATA_SCIENCE\PROJECT\MINI_PROJECT_2\Student_Performance_Analysis.xlsx'

marks_df = pd.read_excel(file_path, sheet_name='Marks')
attendance_df = pd.read_excel(file_path, sheet_name='Attendance')

marks_df.head(100), attendance_df.head(100)

```

### 2.Merge the two DataFrames on the Nmae column to create a Single DataFrame.

```python
student_df = pd.merge(marks_df,attendance_df,on='Name',how='inner')

student_df.head(15)
```

### 3. Replace Attendance values ('Y'or 'N') with numeric Values (1 for'Y' and 0 for 'N').

```python
attendance_cols = attendance_df.columns.drop('Name')

student_df[attendance_cols] = student_df[attendance_cols].replace({'Y': 1,'N': 0})

attendance_cols
```

```python
student_df[attendance_cols]
```

### 4. Deal with the missing values.

```python
student_df.isna().sum()
```

```python
numeric_cols = student_df.select_dtypes(include='number').columns
student_df[numeric_cols] = student_df[numeric_cols].fillna(student_df[numeric_cols].mean())

student_df[numeric_cols]
```

```python
categorical_cols = student_df.select_dtypes(include='object').columns
for col in categorical_cols:
    student_df[col] = student_df[col].fillna(student_df[col].mode()[0])

student_df[col]    
```

### 5. Do the column cleaning like name column os the combination of upper and lower case.

```python
student_df['Name'] = student_df['Name'].str.strip().str.title()

student_df['Name'] = student_df['Name'].str.replace(r'\s+', ' ', regex=True)

student_df['Name']
```

### Attendance Percentage for each student based on attendance columns.

```python
attendance_cols=[
    col for col in student_df.columns
    if col.startswith('Attendance')
]    
```

### Calculate the weighted percentage – Attendance(40%), Mini test01(10%), Mini test02(10%), Live test(20%), Assignment (20%).

```python
student_df['weighted_%']=(student_df['Attendance_%']*0.40 + student_df['Mini Test 1'] * 0.10 + student_df['Mini Test 2'] * 0.10 + student_df['Live Test'] * 0.20 + student_df['Assignment']*0.20)
```

### 1.	Create a bar chart displaying weighted percentages for top 5 students.

```python
top_5 = student_df.sort_values(
    'weighted_%', ascending=False).head(5)

import matplotlib.pyplot as plt

plt.figure(figsize=(15,4))
plt.bar(top_5['Name'], top_5['weighted_%'])
plt.xlabel('Student Name')
plt.ylabel('Weighted Percentage')
plt.title('Top 5 Students by Weighted Percentage')
plt.xticks(rotation=30)
plt.show()
```

```python
def classify_performance(w):
    if w >= 85:
        return 'Excellent'
    elif w >= 70:
        return 'Good'
    elif w >= 50:
        return 'Average'
    else:
        return 'Poor'

student_df['Performance'] = student_df['weighted_%'].apply(classify_performance)
```

### 2.	Create a pie chart showing the distribution of students across the four performance categories.

```python
performance_counts = student_df['Performance'].value_counts()

import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.pie(performance_counts,labels=performance_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Student Performance Distribution')
plt.show()

```

### 3.	Create box plots for each test (Live Test, Mini Test 1, Mini Test 2, Assignment) to visualize the spread and detect potential outliers in scores.

```python
import seaborn as sns

test_cols = ['Mini Test 1', 'Mini Test 2', 'Live Test', 'Assignment']

plt.figure(figsize=(8,5))
sns.boxplot(data=student_df[test_cols])
plt.title('Score Distribution Across Tests & Assignment')
plt.ylabel('Marks')
plt.show()
```
